```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
folder = 'test'
df = pd.read_csv(f'{folder}_labels.csv')
df.head()
```

```python
filesname = df['filename']
new_filesname = []

for name in filesname:
    nm = name.replace('ppm','jpg')
    new_filesname.append(nm)
    
new_filesname = pd.Series(new_filesname, name='filename')
```

```python
new_filesname
```

```python
df['filename']=new_filesname
df.head()
```

```python
df.to_csv(f'{folder}_labels_bis.csv', index=False)
```
